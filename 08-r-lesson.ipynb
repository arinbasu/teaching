```r
## we will now learn about functions
# R is based on objects, data, and functions to manipulate the objects and data
# Functions: instructions to conduct procedures in R
# Here is a function that takes an argument x and increases it by 2

increase_by_two <- function(x){
    x2 = x + 2
    return(x2)
}

# note:
# 1. Write the name of the function first. The function name should be something that suggests some action 
# 2. Then put an assignment operator "<-" to store the output of the function to the function name
# 3. then write the word "function" to signal that this is a function
# 4. then put the 'arguments' within a pair of parentheses. This argument can be a single, or more than one argument
# 5. then put the logic of the function within a pair of curly brackets
# 6. Write the expressions of the function
# 7. Finally, the output of the function uses another function within, that is "return()"
# 8. return() results in the final output of the function
# Illustration:
# "increase_by_two" is the name of the function, using this function you will take an argument x and increase its value by 2
# the arrow to the left stores the output of the resulting function to this object
# the logic is that, x2 equals x plus 2 and return(x2)
# you can call the function wherever you like using function_name(argument), so
# if you want to call function increase_by_two, you will do: increase_by_two(10), if you want to increase the value of 10 by 2

print(increase_by_two(100)) # should print 102 to the console

# Can you write a function to convert temperatures given in degree centigrades to fahrenheit?
# use the following formula: f = 1.8 * c + 32 where f = fahrenheit, and c = temperature in centigrade
# what will you do?
# test the function by setting c to 100, f should be 212
```

```r
# How to draw graphs with R
# We will use a package named "ggplot" to draw graphs
# we will load the package first using tidyverse
# if you have not already installed the package, then install first using
# install.packages("tidyverse") and then load by:
# library(tidyverse)
# we will use a data set mpg "miles per gallon" data set to draw graphs
# Question: "Do cars with larger engines burn more fuels?"
# Here's the data set

# head(mpg)

# we will study displ = the car's engine size in litre
# hwy = highway miles
# cty = city milage

# Create a ggplot


## ggplot(data = mpg) +
# geom_point(mapping = aes(x = displ, y = hwy))

# What the above code does
# ggplot(data = mpg) creates a blank canvas with the data mpg in it
# line two geom_point() adds a layer of points to it
# geom_X whatever the Xis, here point, always takes the argument mapping, and 
# mapping ALWAYS pairs with aes() or aesthetic function, and aes() takes
# X and Y axis and other arguments, here we have given display and highway milage
# aes() takes visual properties, like size and shape of the points, axes, etc
# let's a add a third variable, "drive (drv)" to the aesthetics
# depending on the level (so you can change colours, shape, size)
# geom = geometrical object

ggplot(mpg) +
 geom_point(aes(displ, hwy, alpha = drv), colour = "black")

# in the above plot we set the aesthetics with x and y axis, and
# specified what transparency levels we want for drive (drv)
# we also specified that all points must be black
# remember to put the + sign at the end of the line

```

```r
# Split up the data using facets and graph the subsets
ggplot(mpg) +
 geom_point(aes(displ, hwy)) + 
 facet_wrap(~class, ncol = 3)

## what's going on
# 1. First line calls the data
# 2. Second line asks to draw a point plot and set the aesthetics to be x = displ, y = hwy
# 3. then split up the plots based on the class of the vehicle
# 4. facet_wrap() function takes a categorical variable as argument
# 5. the ~ (tilde) suggests that you use a formula for the data
 
```

```r
## different geoms
ggplot(mpg) +
 geom_point(aes(displ, hwy, colour = drv)) +
 geom_smooth(aes(displ, hwy, colour = drv, linetype = drv))
# what's going on
# 1. opened the canvas with mpg data
# 2. point plotted the scatter between engine size and highway milage, based on their drivetrain
# 3. superimposed a smoothing function on the plot
# 4. further divided the lines based on their drivetrain
```

```r
# Make the code more efficient and compact
ggplot(mpg, aes(displ, hwy)) +
 geom_point(aes(colour = class)) +
 geom_smooth(
     data = filter(mpg, class == "subcompact"),
     se = F
 ) + 
 geom_smooth(
 data = filter(mpg, drv == "r"),
     colour = "red",
     se = F
 )

# 1. pass the data upfront and the aesthetics right in the ggplot() 
# 2. keep adding geometry as layers
# 3. you can get subsetted views of the data if you want
```
